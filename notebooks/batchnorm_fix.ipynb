```python
import torch
import torchvision
```

```python
def compare_models(model_1, model_2):
    models_differ = 0
    for key_item_1, key_item_2 in zip(model_1.state_dict().items(), model_2.state_dict().items()):
        if torch.equal(key_item_1[1], key_item_2[1]):
            pass
        else:
            models_differ += 1
            if (key_item_1[0] == key_item_2[0]):
                print('Mismtach found at', key_item_1[0])
            else:
                raise Exception
    if models_differ == 0:
        print('Models match perfectly! :)')
```

```python
def weights_to_array(model):
    model_weights = []
    for param in model.parameters():
        model_weights.append(param) # check without .data
    return model_weights
```

```python
model1 = torchvision.models.resnext50_32x4d(pretrained=False, progress=True)
torch.save(model1, '../model.pt')
```

```python
weights = weights_to_array(model1)
agg_model = torch.load('../model.pt')
for idx, param in enumerate(agg_model.parameters()):
        param.data = weights[idx]
```

```python
compare_models(model1, agg_model)
```
